```python
# import dependencies
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('profilepic.jpeg')
```

```python
# crop the image
img = img[150:801, 100:501]

# convert the image t
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # reorder the color

# show image
plt.imshow(img)
```

```python
# crop the bottom right corner
crop = img[-3:, -3:]
plt.imshow(crop)
```
