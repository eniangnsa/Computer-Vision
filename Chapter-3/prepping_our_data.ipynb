```python
# import necessary packages
from torchvision import datasets
import torch
```

```python
data_folder = '~/data/FMNIST'
fmnist = datasets.FashionMNIST(data_folder, download=True, train=True)
```

```python
tr_images = fmnist.data
tr_targets = fmnist.targets
```

```python
# inspect the tensors
unique_targets = tr_targets.unique()
print(f"The shape of the training data is: {tr_images.shape}, the unique targets are: {unique_targets}")
```

```python
fmnist.classes
```

We have  60,000 images each of size 28x28 and having 10 classes to predict from.

```python
import matplotlib.pyplot as plt 
%matplotlib inline
import numpy as np
```

```python
R, C = len(tr_targets.unique()), 10
fig, ax = plt.subplots(R, C, figsize=(10,10))
for class_label, plot_row in enumerate(ax):
    label_x_rows = np.where(tr_targets == class_label)[0]
    for plot_cell in plot_row:
        plot_cell.grid(False); plot_cell.axis('off')
        ix = np.random.choice(label_x_rows)
        x, y = tr_images[ix], tr_targets[ix]
        plot_cell.imshow(x, cmap = 'gray')
    plt.tight_layout()
```
