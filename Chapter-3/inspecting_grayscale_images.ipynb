```python
# import dependencies
import cv2
import matplotlib.pyplot as plt
```

#### Convert the image into into structured arrays and scalers

The cv2.imread() will help in creating an array of pixel values from the images.

We can verify if this array matches the images by using the plt.imshow() that takes in as argument the array of the image and plots the image

```python
img = cv2.imread('profilepic.jpeg')
```

```python
plt.imshow(img)
```

Next, let's crop the image between the 150th-800th row, as well as 100th-500th columns.

Then convert it to grayscale and plot the output

```python
# crop the image
img = img[150:801, 100:501]

# convert the image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# show image
plt.imshow(img_gray)
```

```python
img.shape
```

```python
# convert the image into a 25 x 25  array and plot
img_gray_small = cv2.resize(img_gray, (25, 25))
plt.imshow(img_gray_small)
```
